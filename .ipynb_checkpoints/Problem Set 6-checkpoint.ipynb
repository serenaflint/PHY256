### Problem 1

```python
import numpy as np
import numpy.linalg as LA

def norm(masses, idx1, idx2, k):
    #assertion for my own sanity in testing code
    assert(idx1.size == idx2.size == k.size), "Spring parameter arrays have unequal sizes"
    
    M = np.diag(masses)
    A = np.zeros((k.size+1, k.size+1))
    
    for i in range (0,idx1.size):
        A[idx1[i],idx2[i]] += k[i]
        A[idx2[i],idx1[i]] += k[i]
        A[idx2[i],idx2[i]] -= k[i]
        A[idx1[i],idx1[i]] -= k[i]
        
    M_inv = LA.inv(M)
    B = np.dot(M_inv, A)
    
    print("\nM:\n", M)
    print("\nM Inverse:\n", M_inv)
    print("\nA:\n", A)
    print("\nB:\n", B)
    
    eig_val, eig_vec = LA.eig(B)
    exp_omg = np.sqrt(np.abs(eig_val))
    
    print("\nCalculated Eigenvalues:",eig_val)
    print("\nCalculated Omegas:", exp_omg)
    
    #eig_val, eig_vec = LA.eig(np.dot(LA.inv(M),A)) #one-liner of the above process
    return eig_val

#testing with sample values!
#masses = 2*np.ones(5)
#idx1 = np.array([0,1,2,3])
#idx2 = np.array([1,2,3,4])
#k = np.ones(4)
#print("\nCalculuated Eigenvalues:", norm(masses, idx1, idx2, k))
```

### Problem 2

```python
N = 2
masses = np.ones(N)
idx1 = np.array([0])
idx2 = np.array([1])
k = 2*np.ones(N-1)

norm(masses, idx1, idx2, k)

k_p = 2
mtot_p = 1.0/2.0 #(1*1)/(1+1)
pred = np.sqrt(k_p/mtot_p)
print("\nExpected Angular Frequencies: [",0, " ", pred, "]")
#print("\nExpected Omegas:",np.sqrt(np.abs(norm(masses, idx1, idx2, k))))
```

We see a slight discrepancy with the first eigenvalue/calcuated omega with the first expected angular frequency. This appears to be entirely due rounding errors within python, as the first calculated eigenvalue is on the order of $10^{-16}$.

### Problem 3

```python
import matplotlib.pyplot as plt
%matplotlib inline  
%config InlineBackend.figure_format = 'retina'

#initial index and mass arrays
N = 100
masses = 2*np.ones(N)
idx1 = np.zeros(N-1, dtype="int")
idx2 = np.ones(N-1, dtype="int")
x = np.ones(N)

#filling the index arrays
for i in range(0,N-1):
    idx1[i]+= i
    idx2[i]+= i
    x[i]+= i

x[N-1] = N
#print(idx1)
#print(idx2)

k = np.ones(N-1)
y = norm(masses, idx1, idx2, k)
y.sort()
print("\nEigenvalues:\n", y)

plt.figure(figsize=(9,6))
plt.plot(x, y, 'b-')
plt.title("Normal Mode Frequencies, $N=100$")
plt.ylabel("Eigenvalues")
plt.xlabel("Node")
```
