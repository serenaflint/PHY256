```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from math import erfc

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

def walk(steps):
    r = stats.norm.rvs(size=steps)
    return sum(r)

def nparticle_walk(particles,steps):
    x_vec = np.zeros(0)
    for i in range(particles):
        x = walk(steps)
        x_vec = np.append(x_vec, x)
    return x_vec

def nparticle_walk_ditch(particles, steps, x0):
    x_vec = np.zeros(0)
    for i in range(particles):
        x = walk(steps)
        if(x <= x0):
            x_vec = np.append(x_vec, x)
    return x_vec
            
```

#### Part 1

```python
steps = 100
particles = 1000
x0 = 15

x_vec = nparticle_walk(particles, steps)

n, bins, patches = plt.hist(x_vec, bins=25, density=1, color='olivedrab')
plt.title("Random Walks, No Ditch")
plt.xlabel("x", fontsize=14)
plt.ylabel("$p(x)$", fontsize=14)

gauss = stats.norm(0.0, np.sqrt(steps))
y = gauss.pdf(bins)
plt.plot(bins, y, 'k-')
```

```python
steps2 = 100
x_vec_ditch = nparticle_walk_ditch(particles, steps2, x0)

n2, bins2, patches2 = plt.hist(x_vec_ditch, bins = 25, density=1, color='palevioletred')
plt.title("Random Walks, with Ditch N=100, x0=15")
plt.xlabel("x", fontsize=14)
plt.ylabel("$p(x)$", fontsize=14)

gauss2 = stats.norm(0.0, np.sqrt(steps2))
y2 = gauss2.pdf(bins2)
plt.plot(bins2, y2, 'k-')
```

Slightly lopsided!

```python
def frac_ditch(steps, x0):
    inside = x0/(np.sqrt(2.0)*np.sqrt(steps))
    return 0.5*erfc(inside)

particles6 = 1000
steps6 = 60
x06 = 10

percent_calc = frac_ditch(steps6,x06)*100
print("Percent Calculated (N=60, x0=10):", percent_calc, "%")
```

#### Part 1

```python
import os

def p(x):
    a = 1.0/np.sqrt(2.0*np.pi)
    b = np.exp(-x**2/2.0)
    return a*b

def f(y, fwhm):
    a = 1.0/(np.pi*fwhm)
    b = fwhm**2/(y**2+fwhm**2)
    return a*b

def z(x, y, fwhm):
    return p(x)*f(y, fwhm)

def rand_walk(N_steps, fwhm):
    x_arr = np.zeros(N_steps)
    y_arr = np.zeros(N_steps)

    sign = 2*np.random.randint(0,2, size=N_steps)-1
    #print(sign) #debugging

    for i in range(0,N_steps):
        x = stats.norm.rvs()
        x_arr[i] = x
        y = stats.cauchy.rvs()
        y_arr[i] = y
    
    #print(x_arr) #debugging
    #print(y_arr) #debugging

    z_arr = z(x_arr, y_arr, fwhm)
    z_new = z_arr*sign

    for i in range(1,N_steps):
        z_new[i] += z_new[i-1]

    #print(z_arr) #debugging
    return z_new
    
```

#### Part 2

```python
fwhm = 1.0
N_steps = 10000

z_plot = rand_walk(N_steps, fwhm)

x_plot = np.ones(N_steps)
for i in range(0,N_steps):
    x_plot[i]+= i

plt.figure(figsize=(9,6))
plt.title("1D Random Walk, $\gamma = 2$")
plt.ylabel("p(x) + f(y)")
plt.plot(x_plot, z_plot, '-', color='teal')
```

```python
n7, bins7, patches7 = plt.hist(z_plot, bins = 60,density=1, color="powderblue")
plt.title("Z-Value Distribution from Random Walks, $\gamma = 1.0$")
plt.ylabel("$p(x)$", fontsize=14)

cauchy7 = np.linspace(stats.cauchy.ppf(0.01), stats.cauchy.ppf(.99), 100)
y7 = stats.cauchy.pdf(bins7)
plt.plot(bins7, y7, 'k-')
```

```python
from collections import deque
z_queue = deque(z_plot2) #CSC 172/data structures is finally relevant!
std_plot = np.zeros(N_steps)
z_temp = [] #empty list

for i in range(0,N_steps):
    zi = z_queue.popleft() #removes first added element
    z_temp.append(zi) #adds it to a list, using a list keeps it the necessary size
    stdev = np.std(z_temp) #takes stdev of the list
    std_plot[i] = stdev #adds value to the y-axis array
    
    
plt.figure(figsize=(9,6))
plt.title("Standard Deviation vs Time Step")
plt.ylabel("$\sigma$", fontsize = 16)
plt.xlabel("Time Step", fontsize = 16)
plt.plot(x_plot2, std_plot, '-', color='y')
```
